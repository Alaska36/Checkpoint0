```python
import math

class Point3D:
    def __init__(self, x, y, z):
        self.x = x
        self.y = y
        self.z = z
    
    def get_coordinates(self):
        return self.x, self.y, self.z

# Créer une instance de Point3D
my_point = Point3D(1, 2, 3)
print("Coordinates of my_point:", my_point.get_coordinates())

# Classe Rectangle
class Rectangle:
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    def area(self):
        return self.length * self.width
    
    def perimeter(self):
        return 2 * (self.length + self.width)

# Créer une instance de Rectangle
my_rectangle = Rectangle(length=4, width=3)
print("Area of my_rectangle:", my_rectangle.area())
print("Perimeter of my_rectangle:", my_rectangle.perimeter())

# Classe Circle
class Circle:
    def __init__(self, center_x, center_y, radius):
        self.center_x = center_x
        self.center_y = center_y
        self.radius = radius
    
    def area(self):
        return math.pi * self.radius**2
    
    def perimeter(self):
        return 2 * math.pi * self.radius
    
    def inside(self, x, y):
        distance = math.sqrt((x - self.center_x)**2 + (y - self.center_y)**2)
        return distance <= self.radius

# Créer une instance de Circle
my_circle = Circle(center_x=0, center_y=0, radius=5)
print("Area of my_circle:", my_circle.area())
print("Perimeter of my_circle:", my_circle.perimeter())
print("Is (2, 2) inside my_circle:", my_circle.inside(2, 2))

# Classe Bank
class Bank:
    def __init__(self, balance=0):
        self.balance = balance
    
    def deposit(self, amount):
        self.balance += amount
        return self.balance
    
    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            return self.balance
        else:
            return "Insufficient funds"
        
# Créer une instance de Bank
my_bank = Bank(balance=1000)
print("Current balance:", my_bank.balance)
print("New balance after deposit:", my_bank.deposit(500))
print("New balance after withdrawal:", my_bank.withdraw(300))
```
